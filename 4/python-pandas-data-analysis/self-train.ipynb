```python
# converting the date coloumn into a date datatype for proper
# Date and time manipulation
covid_rec_df['date']
# currenty the datatype is an object so we will convert it into
# a date type coloumn
```

```python
covid_rec_df['date']=pd.to_datetime(covid_rec_df.date)
covid_rec_df['date']
# Now the date coloumns datatype has been converted to 
# datetime64
```

```python
temp_covid_rec_df=covid_rec_df.copy()
temp_covid_rec_df['month']=pd.DatetimeIndex(temp_covid_rec_df['date']).month
temp_covid_rec_df[temp_covid_rec_df.month == 5][['new_cases', 'new_tests']]
```

```python
temp_covid_rec_df['day']=pd.DatetimeIndex(temp_covid_rec_df.date).day
temp_covid_rec_df['month']=pd.DatetimeIndex(temp_covid_rec_df.date).month
temp_covid_rec_df['year']=pd.DatetimeIndex(temp_covid_rec_df.date).year
temp_covid_rec_df['weekday']=pd.DatetimeIndex(temp_covid_rec_df.date).weekday
temp_covid_rec_df
```

```python
temp_covid_rec_df[temp_covid_rec_df['month'] == 5][['new_deaths', 'new_tests', 'new_cases']].sum()
```

```python
temp_covid_month_df=temp_covid_rec_df.groupby('month')[['new_cases']].sum()
temp_covid_month_df
```
