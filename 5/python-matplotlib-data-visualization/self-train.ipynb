```python
# First let's plot graphs of arrays
fruit_yields = [3.2, 1.4, 2.4, 5.3, 6.5, 4.1]
```

```python
plt.plot(fruit_yields)
```

```python
plt.plot(fruit_yields);
# Use ';' at the end to prevent any additional outputs
```

### Customizing the axis and its labels

```python
years = [2010, 2011, 2012, 2013, 2014, 2015]

plt.plot(years, fruit_yields);
```

```python
plt.xlabel('Year')
plt.ylabel('Yields')
plt.plot(years, fruit_yields)
```

```python
vegetable_yields = [3.4, 5.5, 1.2, 2.4, 5.6, 6.47]

plt.plot(years, fruit_yields, marker='o')
plt.plot(years, vegetable_yields, marker='s')

plt.xlabel('Years')
plt.ylabel('Yields')

plt.title('Yields of fruits and vegetables')
plt.legend(['Fruits', 'Vegetables'])
```
